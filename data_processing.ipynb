as we checked before about distribution, we've come into conclusion into using IQR for which we need to follow the steps below as there's not normally distributed

1. find Q1 and Q3
2. IQR = Q3 - Q1
3. Find lower and upper limit
4. lower_limit = Q1 - 1.5 * IQR
5. upper_limit = Q3 + 1.5 * IQ6
6. Find the records/ values / datapoints greater than upper limit and less than lower (outliers)
7. find the percentage of outliers 
8. if the percentage is less than 5% then only we will handle the outliers
9. if the percenatge is greater the 5% then do not handle

```python
for column in numerical_columns.drop('PerformanceRating'):
    # find Q1 and Q3
    Q1 = data_without_target[column].quantile(0.25)
    Q3 = data_without_target[column].quantile(0.75)

    #IQR
    IQR = Q3 - Q1

    # find lower and upper limit
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR


    # find the values that are outside of upper and lower limit
    outliers = data_without_target.loc[(data_without_target[column] > upper_limit) | (data_without_target[column] < lower_limit)]

    # get the percentage of outliers
    percent_outlier = (len(outliers)/ len(data_without_target))*100

    print(f'percentage outlier --> {column}: {percent_outlier}')

    if 0 < percent_outlier  <= 5:
        outliers = data_without_target[column].median()
        print("""
-------------------------------
              outlier handeled
              -------------------------
""")
```

### Encoding for Categorical Column

```python
for col in data_without_target.columns:
    if data_without_target[col].dtype == 'O':
        # Get unique values after replacement
        unique_values = data_without_target[col].unique()
        print(f"{col}'s values : {unique_values}\n")
```

using pandas category.codes option to convert categorical into numerical values

```python
for col in data_without_target.columns:
    if data_without_target[col].dtype == 'O':
        # Convert object-type columns to categorical dtype
        data_without_target[col] = data_without_target[col].astype('category')
        # Replace categorical values with numerical codes
        data_without_target[col] = data_without_target[col].cat.codes

data_without_target
```

### Scaling Data

We'll use **MinMaxScaler** here. Main reason to use minmaxscaler is in here maximum features are not normally distributed here mostly are right skewd. Since Standard Scaler is useful where features are normally distributed, we didn't use standard scaler for this reason

```python
# Initialize the scaler
scaler = MinMaxScaler()

# Scale the training data
data_without_target_scaled = data_without_target.copy()
data_without_target_scaled[data_without_target_scaled.columns] = scaler.fit_transform(data_without_target_scaled)
data_without_target_scaled
```

```python
preprocessed_data = pd.concat([data_without_target, df['PerformanceRating']], axis=1)
preprocessed_data_scaled = pd.concat([data_without_target_scaled, df['PerformanceRating']], axis=1)

data_randomized = preprocessed_data.sample(frac=1, random_state=42)  # Randomize with a fixed random state for reproducibility
data_randomized_scaled = preprocessed_data_scaled.sample(frac=1, random_state=42)

# Split the dataset into training (80%) and testing (20%) sets
train_validation_data, test_data = train_test_split(data_randomized, test_size=0.2, random_state=42)
train_validation_data_scaled, test_data_scaled = train_test_split(data_randomized_scaled, test_size=0.2, random_state=42)

# Save the training and testing sets to CSV files
train_validation_data.to_csv('../../data/processed/train_validation_data.csv', index=False)  # Save training set to CSV without index
test_data.to_csv('../../data/processed/test_data.csv', index=False)

# Save the scaled training and testing sets to CSV files
train_validation_data_scaled.to_csv('../../data/processed/train_validation_data_scaled.csv', index=False)
test_data_scaled.to_csv('../../data/processed/test_data_scaled.csv', index=False)
```
