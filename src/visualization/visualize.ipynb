# Visualize Data

Here we try explore all data in terms of visualisation only which can gives us insights. We have already made all possible insights in "data_exploratory_analysis" part. For which we'll not go in detail regarding insights part rather than we'll try to focus on visualisation part only where we'll explore more visualisation which didn't cover in "data_exploratory_analysis" part.

```python
# Import nessecary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Define the relative path to the file
file_path = '../../data/processed/INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.csv'

# Read the file into a pandas DataFrame
df = pd.read_csv(file_path)
df
```

Relationship between Age and TotalWorkExperienceInYears

```python
sns.lineplot(x='Age',y='TotalWorkExperienceInYears',data=df)
plt.xlabel('Age',fontsize=15)
plt.ylabel('TotalWorkExperienceInYears',fontsize=15)
plt.show()
```

Relation between Employee last salary hike and number of company worked

```python
sns.boxplot(x='EmpLastSalaryHikePercent',y='NumCompaniesWorked',data=df)
plt.xlabel('EmpLastSalaryHikePercent',fontsize=15)
plt.show()
```

Relation between Years Since Last Promotion and Experience Years In CurrentRole

```python
sns.lineplot(x='YearsSinceLastPromotion',y='ExperienceYearsInCurrentRole',data=df)
plt.xlabel('YearsSinceLastPromotion',fontsize=10)
plt.show()
```

Relation between Distance From Home and Employee Last Salary Hike Percent

```python
sns.lineplot(x='DistanceFromHome',y='EmpLastSalaryHikePercent',data=df)
plt.xlabel('DistanceFromHome',fontsize=15)
plt.show()
```

Relationship between Gender & Number companies worked with respect to performance rating

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Gender',y='NumCompaniesWorked',hue=df.PerformanceRating,data=df)
plt.xlabel('Gender',fontsize=20)
plt.ylabel('NumCompaniesWorked',fontsize=15)
plt.show()
```

Relationship between Attrition and years with current manager with respect to performance rating

```python
plt.figure(figsize=(20,10))
sns.violinplot(x='Attrition',y='YearsWithCurrManager',hue=df.PerformanceRating,data=df)
plt.xlabel('Attrition',fontsize=20)
plt.ylabel('YearsWithCurrManager',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Age',y='TotalWorkExperienceInYears',hue=df.PerformanceRating,data=df)
plt.xlabel('Age',fontsize=20)
plt.ylabel('TotalWorkExperienceInYears',fontsize=15)
plt.show()
```

```python
plt.subplots(figsize=(15, 5))
bar = sns.barplot(data = df, x= 'EmpDepartment', y = 'PerformanceRating',hue= 'EmpEnvironmentSatisfaction',
                  palette = 'magma').set(title = "Performance rating vs Environment satisfaction")
```

```python
plt.subplots(figsize=(10, 5))
kde = sns.kdeplot(data=df, x="EmpLastSalaryHikePercent", y ='PerformanceRating',multiple="stack",
                  palette = 'viridis', fill = True, linewidth = 0)
kde.set(title = 'Relation between Salary hike & Performance rating')
```

```python
# Create a count plot
plt.figure(figsize=(10, 6))
sns.countplot(y="PerformanceRating", hue="EmpLastSalaryHikePercent", data=df)
plt.title("Performance Rating Count by Salary Hike Percent")
plt.xlabel("Count")
plt.ylabel("Performance Rating")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Age', y='PerformanceRating')
plt.title("lineplot of Age vs Performance Rating")
plt.xlabel("Age")
plt.ylabel("Performance Rating")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='TotalWorkExperienceInYears', y='PerformanceRating')
plt.title("lineplot of experience vs Performance Rating")
plt.xlabel("experience")
plt.ylabel("TotalWorkExperienceInYears")
plt.show()
```
